# Coalesce/Dealing with Missing Values

```python
import pandas as pd
import numpy as np

df2 = pd.DataFrame({
    'Hourly Rate': [20, 30, np.nan, 28, np.nan, 17, np.nan,22, 32],
    'Daily Rate': [72, 74,65,80, 74, np.nan, 67,82,75]})
df2
```

## Mask Function

```python
# [New Column]   [using Column values]  [<- Values masked] [NULL values] , [using Column values]
df2['Final Rate']=df2['Hourly Rate'].mask(pd.isnull, df2['Daily Rate'])
df2
```

## Using Numpy 

```python
df2['Final Rate'] = np.where(df2['Hourly Rate'].isnull(),df2['Daily Rate'],df2['Hourly Rate'])
```

```python
# fill with values in next column
df3.copy().bfill(axis=1)
```

```python
# fill with values in row above
df3.copy().ffill(axis=0)
```

# Conditional Functions

Write a program that prints the number for 1 to 50, for number multiple of 2 print ‘fizz’ instead of a number, for numbers multiple of 3 print ‘buzz’, for numbers which are multiples of both 2 and 3 ‘fizzbuzz.’

```python

def Fizz_buzz():
    for i in range(1,51):
        if i%2 == 0 and i%3 == 0:
            print('fizzbuzz')
        elif i%2 == 0:
            print('fizz')
        elif i%3 == 0:
            print('buzz')
        else:
            print (i)
```

```python
Fizz_buzz()
```

# First Unique Character in a String

- Given a string, find the first non-repeating character in it and return its index.
- If it doesn’t exist, return -1.

```python
def first_unique_character(string):
    for i in range(len(string)):      # create a range item
        c = string[i]                 # create a new object for each element in string
        if string.count(c)==1:        # this is the tricky part: for each element within the string, if the count is 1.
            return i                  # return i; 
                                      # we can also use the index method: return string.index(c)
        return -1

string = "I am a data scientist."
first_unique_character(string)
```

# Reverse String Order

```python
sentence = "I love programming in Python. How about you?"

def reversed_string(string):
    return string[::-1]     #returns string in reverse order


reversed_string(sentence)
```

# Check If Two String Arrays are Equivalent

```python
def two_string_equivalent(word1,word2):
    return "".join(word1) == "".join(word2) # combining list of 

str1 = ["ab", "c"]
str2 = ["a", "bc"]
two_string_equivalent(str1,str2)
```

# In String Find Palindromes (Words spelled the same fowards and backwards)

```python
sentence2 = "To be or not to be a data scientist, this is not a question. Ask your mom, lol."
def find_palindrome(string):
    palindrome = []    # emtpy list
    
    # 1 ask if Case matter
    string = string.lower()
    
    # 2 ask if punctuation & numbers matter 
    import re
    newstring = re.sub(r'\W+', ' ', string)  # replaces all characters not letters with a space {can add numbers with "\D"}
    
    
    # Separate each word in string
    words = newstring.split(' ')
    
    for x in words:
        if x == x[::-1]:
            palindrome.append(x)
            
    return set(palindrome)      
        
find_palindrome(sentence2)
    
```
